```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# BTC
# get historical data
btc = yf.download('BTC-USD', start='2018-12-31', end='2024-01-01') # end is not included
btc.index = pd.to_datetime(btc.index)

btc_volume = btc[['Volume']]
btc = btc[['Close']]

```

```python
df = pd.DataFrame(btc)
df.columns = ['btc']
df['btc_volume'] = btc_volume
```

```python
#plot
btc.plot(figsize=(10, 6), label='BTC-USD')
plt.title(f"BTC-USD Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python

btc_volume.plot(figsize=(10, 6), label='BTC-USD')
plt.title(f"BTC-USD Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# Create the plot figure and axis
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot the 'Close' price on the first y-axis (ax1)
sns.lineplot(data=btc, x=btc.index, y='Close', ax=ax1, color='tab:blue', label='BTC Close Price')

# Label the first y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('BTC Close Price (USD)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create the second y-axis (ax2) to plot 'Volume'
ax2 = ax1.twinx()
sns.lineplot(data=btc_volume, x=btc_volume.index, y='Volume', ax=ax2, color='tab:orange', label='BTC Volume', linestyle='--')

# Label the second y-axis
ax2.set_ylabel('BTC Volume', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Title and formatting
plt.title('Bitcoin Price and Volume Over Time')
plt.xticks(rotation=45)


# Show the plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error
from sklearn.svm import SVR


```

```python
# linear regression
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.3, random_state=42)

model = LinearRegression()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"linear regression: \nrmse: {rmse}, r2: {r2}, mae: {mae}")
```

```python
# decision tree
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.3, random_state=42)

# Initialize the Decision Tree Regressor model
dt_model = DecisionTreeRegressor(random_state=42)

# Fit the model on the training data
dt_model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = dt_model.predict(x_test)

rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f"decision tree: \nrmse: {rmse}, r2: {r2}, mae: {mae}")
```

```python
#random forest
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.3, random_state=42)

model = RandomForestRegressor(random_state=42)

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f"Random forest: \nrmse: {rmse}, r2: {r2}, mae: {mae}")
```

```python
# SVM
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.3, random_state=42)

model = SVR(kernel='linear')

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f"SVM: \nrmse: {rmse}, r2: {r2}, mae: {mae}")
```
