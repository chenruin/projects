```python
# BTC
# get historical data
btc = yf.download('BTC-USD', start='2018-12-31', end='2024-01-01') # end is not included
btc.index = pd.to_datetime(btc.index)

btc_volume = btc[['Volume']]
btc = btc[['Close']]

```

```python
# standardize the variables

df = df[:-1]

y = df['next_close']
x = df.drop(columns=['next_close', 'btc'])

scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)

y_scaled = scaler.fit_transform(y.values.reshape(-1, 1))

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error
from sklearn.svm import SVR

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
```

```python
# linear regression
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.3, random_state=42)

model = LinearRegression()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"linear regression: \nrmse: {rmse}, r2: {r2}, mae: {mae}")
```

```python
# decision tree
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.3, random_state=42)

# Initialize the Decision Tree Regressor model
dt_model = DecisionTreeRegressor(random_state=42)

# Fit the model on the training data
dt_model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = dt_model.predict(x_test)

rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f"decision tree: \nrmse: {rmse}, r2: {r2}, mae: {mae}")
```

```python
#random forest
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.3, random_state=42)

model = RandomForestRegressor(random_state=42)

model.fit(x_train, y_train)

y_pred_rf = model.predict(x_test)

rmse = root_mean_squared_error(y_test, y_pred_rf)
r2 = r2_score(y_test, y_pred_rf)
mae = mean_absolute_error(y_test, y_pred_rf)
print(f"Random forest: \nrmse: {rmse}, r2: {r2}, mae: {mae}")
```

```python
# SVM
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.3, random_state=42)

model = SVR(kernel='linear')

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f"SVM: \nrmse: {rmse}, r2: {r2}, mae: {mae}")
```

```python
# SVM
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.3, random_state=42)

model = SVR(kernel='rbf')

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f"SVM: \nrmse: {rmse}, r2: {r2}, mae: {mae}")
```

```python
y_train = y_train.reshape(-1, 1)  
y_test = y_test.reshape(-1, 1)    

# Convert to TensorFlow tensors if necessary
x_train_tensor = tf.convert_to_tensor(x_train, dtype=tf.float32)
y_train_tensor = tf.convert_to_tensor(y_train, dtype=tf.float32)
x_test_tensor = tf.convert_to_tensor(x_test, dtype=tf.float32)
y_test_tensor = tf.convert_to_tensor(y_test, dtype=tf.float32)

# Define the model using Keras (TensorFlow)
model = models.Sequential()

# Input layer and first hidden layer
model.add(layers.Dense(64, activation='relu', input_shape=(x_train.shape[1],)))

# Second hidden layer
model.add(layers.Dense(32, activation='relu'))

# Output layer (single output for regression)
model.add(layers.Dense(1))

# Compile the model with Mean Squared Error loss for regression
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(x_train_tensor, y_train_tensor, epochs=1000, batch_size=16, validation_data=(x_test_tensor, y_test_tensor), verbose=1)
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss during Training')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.show()
```
